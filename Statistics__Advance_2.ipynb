###6. Define the terms "probability distribution" and "probability density function" (PDF). How does a PDF differ from a probability mass function (PMF)?

Ans:-
Probability Distribution
A probability distribution describes how the probabilities of a random variable are distributed across its possible values. It provides a complete characterization of the random variable, indicating the likelihood of each possible outcome. Probability distributions can be categorized into two main types:

Discrete Probability Distribution: For discrete random variables, where outcomes can be counted (e.g., rolling a die). The probabilities are associated with distinct values.

Continuous Probability Distribution: For continuous random variables, where outcomes can take on any value within a range (e.g., measuring time or height). The probabilities are represented over intervals rather than specific values.

Probability Density Function (PDF)
A probability density function (PDF) is a function that describes the likelihood of a continuous random variable taking on a specific value. The PDF is used to compute probabilities for continuous variables by integrating the PDF over a given interval. Key characteristics of a PDF include:

The value of the PDF at any specific point can be greater than 1; however, the area under the entire curve (the integral across all possible values) must equal 1.
The probability of the random variable falling within a certain interval is found by integrating the PDF over that interval.
Probability Mass Function (PMF)
A probability mass function (PMF) is a function that defines the probability of a discrete random variable taking on specific values. The PMF provides the probabilities associated with each possible value of the discrete random variable. Key characteristics of a PMF include:

The sum of the probabilities across all possible values must equal 1.
Each individual probability is between 0 and 1.
Differences Between PDF and PMF
Type of Variable:

PDF: Used for continuous random variables.
PMF: Used for discrete random variables.
Probability Representation:

PDF: Represents probabilities as areas under a curve. The probability of a specific value is technically 0; instead, probabilities are found over intervals.
PMF: Represents probabilities directly for specific outcomes. Each outcome has a non-zero probability.
Calculation of Probabilities:

PDF: Probabilities are obtained by integrating the PDF over an interval.
PMF: Probabilities are obtained by summing the PMF values for specific outcomes.

Based on a significance level of 0.05, do you reject or fail to reject the null hypothesis? Task: Write Python code to calculate the z-score and p-value for the given data. Objective: Apply the formula for the z-score and interpret the p-value for hypothesis testing.

Ans:- To calculate the z-score and p-value using the given data, we can follow these steps:

Calculate the z-score using the formula:

Calculate the p-value corresponding to the calculated z-score. Since we typically perform a two-tailed test when evaluating if the sample mean is significantly different from the population mean, we will use the cumulative distribution function (CDF) for this purpose.

Interpret the p-value in the context of the significance level (0.05) to decide whether to reject or fail to reject the null hypothesis.

```python
import scipy.stats as stats
import numpy as np

# Given data
sample_mean = 105
population_mean = 100
std_dev = 15
sample_size = 25

# Calculate the z-score
z_score = (sample_mean - population_mean) / (std_dev / np.sqrt(sample_size))

# Calculate the p-value (two-tailed test)
p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

# Output the results
print(f"Z-Score: {z_score:.4f}")
print(f"P-Value: {p_value:.4f}")

# Determine whether to reject or fail to reject the null hypothesis
alpha = 0.05
if p_value < alpha:
    decision = "Reject the null hypothesis"
else:
    decision = "Fail to reject the null hypothesis"

print(f"Decision: {decision}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the binomial distribution
n_trials = 10
p_success = 0.6
n_samples = 1000

# Generate samples from a binomial distribution
samples = np.random.binomial(n_trials, p_success, n_samples)

# Calculate the expected mean and variance
expected_mean = n_trials * p_success
expected_variance = n_trials * p_success * (1 - p_success)

# Print the expected mean and variance
print(f"Expected Mean: {expected_mean}")
print(f"Expected Variance: {expected_variance}")

# Plot the distribution of the samples
plt.figure(figsize=(10, 6))
plt.hist(samples, bins=np.arange(-0.5, n_trials + 1.5, 1), density=True, alpha=0.7, color='blue', edgecolor='black')
plt.title('Binomial Distribution (n=10, p=0.6)')
plt.xlabel('Number of Successes')
plt.ylabel('Probability Density')
plt.xticks(range(n_trials + 1))
plt.grid(axis='y')
plt.show()
```
